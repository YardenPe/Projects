```python
df.set_index('release_date', inplace=True)
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

```python
#convert duration to seconds
df['duration']= df['duration_ms'].apply(lambda x: round(x/1000))
df.drop('duration_ms', inplace=True, axis=1)
```

```python
numeric_df = df.drop(['key', 'mode', 'explicit'], axis=1)
numeric_df = numeric_df.select_dtypes(include='number')
corr_df = numeric_df.corr()

plt.figure(figsize=(14,6))
heatmap = sns.heatmap(data=corr_df, annot=True)
heatmap.set_title('Correlation Heatmap')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
sample_df = df.sample(int(len(df)*0.005))
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='loudness', x='energy', color='c').set(title='Loudness Vs. Energy Correlation')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='popularity', x='acousticness', color='b').set(title='Loudness Vs. Energy Correlation')
plt.show()
```

```python
df['dates']= df.index.get_level_values('release_date')
df['dates']= pd.to_datetime(df.dates, format='mixed')
years= df['dates'].dt.year
```

```python
sns.displot(years, kind='hist',discrete=True,aspect=2,height=5).set(title='Number Of Songs Per Year')
plt.show()
```

```python
sns.color_palette("husl")
fig, ax = plt.subplots(figsize=(18,7))
sns.barplot(x=years, y=df['duration'], ax=ax, err_kws={'linewidth': False}, palette="husl")
ax.set(title='Year Vs. Duration')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.set_style(style= "whitegrid")
fig, ax = plt.subplots(figsize=(10, 5))

sns.lineplot(x=years, y=df['danceability'], ax=ax, label='Danceability')
sns.lineplot(x=years, y=df['energy'], ax=ax, label='Energy', color='r')

ax.set(title='Year vs. Danceability and Energy')
plt.xticks(rotation=90)
ax.legend()
plt.show()
```
