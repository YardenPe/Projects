## Scatter Plot

```python
books = pd.read_csv("amazon.csv")
books
```

```python
fig = px.scatter(books, x="Price",y="Reviews",
            title="Number of Reviews by Book Price<br><sup>Amazon Bestseller List 2009 to Present</sup>",
            labels={"Reviews": "# of Reviews", "Price": "Price (USD)"},
            color="Genre",
            template="ggplot2")
fig.show()
```

```python
fig = px.scatter(books, x="Price",y="Reviews", opacity=0.5)
fig.show()
```

## Geographical Scatter Plot

```python
heavy =pd.read_csv("./heavy_meteorites.csv")
heavy
```

```python
fig = px.scatter_geo(heavy,
                    lat= "reclat",
                    lon="reclong",
                    title="Meteorite Landings")
fig.show()
```

```python
fig = px.scatter_geo(heavy,
                     lat="reclat",
                     lon="reclong",
                     title="Meteorite Landings",
                     color="fall",
                     template="plotly_white",
                     size="mass",
                     hover_data=["name", "year"]
                     )
fig.show()
#fall - found in the ground or just seen
#size - scale by the mass in an area
#hover_data - which information to show when hover over each point
```
